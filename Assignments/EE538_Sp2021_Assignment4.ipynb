The drain current of an $n$-channel JFET can be described as a function of $V_{gs}$ by

\begin{equation}
I_d = \beta(V_{gs} - V_{th})^2
\end{equation}

\begin{equation}
g_m = 2\cdot\sqrt{\beta I_D}
